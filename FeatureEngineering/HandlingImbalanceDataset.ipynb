### UPSAMPLING & DOWNSAMPLING

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n_samples=1000
classratio=0.9
class0 = int(n_samples*classratio)
class1 = n_samples-class0
```

```python
### Creating a Dataset
pd1 = pd.DataFrame({
    "feature1": np.random.normal(loc=0,scale=1,size=class0),
    "feature2":np.random.normal(loc=0,scale=1,size=class0),
    "target":[0]*class0
})
### Creating a Dataset
pd2 = pd.DataFrame({
    "feature1": np.random.normal(loc=0,scale=1,size=class1),
    "feature2":np.random.normal(loc=0,scale=1,size=class1),
    "target":[1]*class1
})
df = pd.concat([pd1,pd2]).reset_index(drop=True)
```

```python
df
```

```python
df['target'].value_counts()
```

### UpSampling : Converting the Minority O/P to Major by adding random outputs


```python
dfminor= df[df['target']==1]
dfmajor= df[df['target']==0]
```

```python
from sklearn.utils import resample
dfminor_upsample=resample(dfminor,replace=True,
         n_samples=len(dfmajor),
         random_state=42)
```

```python
dfminor_upsample.head()
```

```python
df_upsample = pd.concat([dfmajor,dfminor_upsample])
df_upsample['target'].value_counts()
```

### Down Sampling: Convert the major o/p into minor

```python
### Creating a Dataset
pd1 = pd.DataFrame({
    "feature1": np.random.normal(loc=0,scale=1,size=class0),
    "feature2":np.random.normal(loc=0,scale=1,size=class0),
    "target":[0]*class0
})
### Creating a Dataset
pd2 = pd.DataFrame({
    "feature1": np.random.normal(loc=0,scale=1,size=class1),
    "feature2":np.random.normal(loc=0,scale=1,size=class1),
    "target":[1]*class1
})
df = pd.concat([pd1,pd2]).reset_index(drop=True)
```

```python
### UpSAMPLING : Creating Artifical Points to create Balance Dataset
dfminor= df[df['target']==1]
dfmajor= df[df['target']==0]
```

```python
from sklearn.utils import resample
dfmajor_downsample=resample(dfmajor,replace=False,
         n_samples=len(dfminor),
         random_state=42)

dfmajor_downsample.shape
```

```python
df_upsample = pd.concat([dfminor,dfmajor_downsample])
df_upsample['target'].value_counts()
```
